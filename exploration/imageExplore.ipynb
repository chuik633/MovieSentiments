```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/image_data.csv')
df.head()
```

```python
def plot_movie_color(movie):
    movie_df = df[df['movie']==movie][['r','g','b']].values/255.0

    fig, ax = plt.subplots(figsize=(8, 1))
    ax.imshow([movie_df], aspect='auto')
    ax.set_axis_off()
    ax.set_title(movie, pad=10)
    plt.show()

plot_movie_color('The Dark Knight')
plot_movie_color('Mad Max: Fury Road')
plot_movie_color("Spirited Away")
plot_movie_color("Interstellar")
```

```python
movies = df["movie"].unique()
fig, axes = plt.subplots(len(movies), 1, figsize=(8, len(movies)*1))

for ax, movie in zip(axes, movies):
    subset = df[df["movie"] == movie][["r", "g", "b"]].values / 255.0
    ax.imshow([subset], aspect="auto")
    ax.set_axis_off()
    ax.set_title(movie, pad=5)

plt.tight_layout()
plt.show()
```
